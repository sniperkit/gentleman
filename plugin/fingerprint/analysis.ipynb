```python
# 初始化环境
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from pymongo import MongoClient

output_notebook()
```

```python
# 数据库
client = MongoClient()
db = client["tgod-test"]
```

## 贴吧帖子列表

```python
threads = db["Thread"].find().sort("create_time")
# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.html#pandas.DataFrame
threads_df = pd.DataFrame(list(threads), columns=["id", "create_time", "last_time", "view_num", "reply_num", "title", "author_id"])
threads_df.set_index("create_time", inplace=True)
threads_df.tz_localize("Asia/Shanghai", copy=False)
threads_df
```

## 贴吧发帖趋势

```python
c = threads_df.resample("M").count()
p = figure()
p.line(c.index, c["id"])
from bokeh.models import DatetimeTickFormatter
p.xaxis.formatter=DatetimeTickFormatter(
        hours=["%d %B %Y"],
        days=["%d %B %Y"],
        months=["%d %B %Y"],
        years=["%d %B %Y"],
    )
show(p)
```
